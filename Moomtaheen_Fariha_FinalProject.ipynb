```python
import numpy as np
import pandas as pd 
from sklearn.model_selection import train_test_split, cross_val_score, RepeatedStratifiedKFold
from sklearn.linear_model import Perceptron

from sklearn.linear_model import LogisticRegression, LinearRegression
from sklearn import svm
from sklearn import tree
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestRegressor
from sklearn.preprocessing import MinMaxScaler, StandardScaler
from sklearn.feature_selection import SelectFromModel, SequentialFeatureSelector, RFECV, RFE

from sklearn.tree import DecisionTreeClassifier
from sklearn.pipeline import Pipeline

from sklearn.metrics import classification_report, confusion_matrix, precision_score, recall_score, f1_score, accuracy_score, mean_squared_error, mean_absolute_error

from sklearn import metrics
import seaborn as sns 
import matplotlib.pyplot as plt 


import warnings 
warnings.filterwarnings("ignore") 
```

```python
df = pd.read_csv("https://corgis-edu.github.io/corgis/datasets/csv/hospitals/hospitals.csv")
df.head()


df.columns
df.shape
```

```python
df.isnull().sum()
```

```python
plt.xlabel("Heart Attack Procedure Cost")
plt.hist(df["Procedure.Heart Attack.Cost"])
df = df[ df["Procedure.Heart Attack.Cost"] > 18000 ]
```

```python
df = df[ df["Rating.Mortality"] != "None"]
df = df[ df["Rating.Safety"] != "None"]
df = df[ df["Rating.Readmission"] != "None"]
df = df[ df["Rating.Experience"] != "None"]
df = df[ df["Rating.Effectiveness"] != "None"]
df = df[ df["Rating.Timeliness"] != "None"]
df = df[ df["Rating.Imaging"] != "None"]
df.shape
```

```python

df = df[df["Facility.Type"] != "Unknown"]

df = df[df["Procedure.Heart Attack.Quality"] != "Unknown"]
df = df[df["Procedure.Heart Attack.Value"] != "Unknown"]

df = df[df["Procedure.Heart Failure.Quality"] != "Unknown"]
df = df[df["Procedure.Heart Failure.Value"] != "Unknown"]

df = df[df["Procedure.Pneumonia.Quality"] != "Unknown"]
df = df[df["Procedure.Pneumonia.Value"] != "Unknown"]

df = df[df["Procedure.Hip Knee.Quality"] != "Unknown"]
df = df[df["Procedure.Hip Knee.Value"] != "Unknown"]

```

```python
X = df.copy()
X = X.drop(columns=["Procedure.Heart Attack.Cost",
       'Procedure.Heart Failure.Quality', 'Procedure.Heart Failure.Value',
        'Procedure.Pneumonia.Quality',
       'Procedure.Pneumonia.Value',
       'Procedure.Hip Knee.Quality', 'Procedure.Hip Knee.Value', 'Procedure.Heart Failure.Cost',
       'Procedure.Pneumonia.Cost', 'Procedure.Hip Knee.Cost'])
cat_cols = ['Facility.Name', 'Facility.City', 'Facility.State', 'Facility.Type',
            'Rating.Mortality', 'Rating.Safety',
       'Rating.Readmission', 'Rating.Experience', 'Rating.Effectiveness',
       'Rating.Timeliness', 'Rating.Imaging',
       'Procedure.Heart Attack.Quality', 'Procedure.Heart Attack.Value']
X = pd.get_dummies(X, columns = cat_cols, drop_first=True)

y = df["Procedure.Heart Attack.Cost"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 42)
```

```python
# L1 regularization
from sklearn.linear_model import LassoCV
lasso = LassoCV()
lasso.fit(X_train, y_train)
selected_features = X.columns[lasso.coef_ != 0]
df_selected = X[selected_features]
df_selected
```

```python
X_train_selected = X_train[selected_features]
X_test_selected = X_test[selected_features]


regressor = LinearRegression()
regressor.fit(X_train_selected, y_train)
# Predictioin
y_pred = regressor.predict(X_test_selected)
# Compute the mean squared error
mse_l1 = mean_squared_error(y_test, y_pred)
print('MSE: {:.2f}'.format(mse_l1))

# mean absolute error
mae_l1 = mean_absolute_error(y_test, y_pred)
print('MAE: {:.2f}'.format(mae_l1))

# root mean squared error
rmse_l1 = mean_squared_error(y_test, y_pred, squared=False)
print('RMSE: {:.2f}'.format(rmse_l1))
```

```python
r_sq_l1 = regressor.score(X_train_selected, y_train)
print('R^2: {:.2f}'.format(r_sq_l1))
print("intercept: ", regressor.intercept_)

print("slope: ", regressor.coef_)
print("Variance score: ", regressor.score(X_test_selected, y_test))
```

```python
# Random Forest feature selection

selector = SelectFromModel(RandomForestRegressor())
selector.fit(X_train, y_train)
selected_features_rf = X.columns[selector.get_support()]
df_selected = X[selected_features_rf]
df_selected
```

```python
X_train_selected = X_train[selected_features_rf]
X_test_selected = X_test[selected_features_rf]


rf = RandomForestRegressor()
rf.fit(X_train_selected, y_train)

# Prediction
y_pred = rf.predict(X_test_selected)
# mean squared error
mse_rf = mean_squared_error(y_test, y_pred)
print("MSE:", mse_rf)

# Compute the mean absolute error
mae_rf = mean_absolute_error(y_test, y_pred)
print('MAE: {:.2f}'.format(mae_rf))

# Compute the root mean squared error
rmse_rf = mean_squared_error(y_test, y_pred, squared=False)
print('RMSE: {:.2f}'.format(rmse_rf))

r_sq_rf = rf.score(X_train_selected, y_train)
print('R^2: {:.2f}'.format(r_sq_rf))
# print("intercept: ", rf.intercept_)

# print("slope: ", rf.coef_)
print("Variance score: ", rf.score(X_test_selected, y_test))
```

```python
# sequential feature selection (forward)

sfs = SequentialFeatureSelector(estimator=LinearRegression(), scoring='neg_mean_squared_error', n_features_to_select='auto')
sfs.fit(X_train, y_train)
selected_features_sfs = X.columns[sfs.get_support()]
df_selected_sfs = X[selected_features_sfs]
df_selected
```

```python
X_train_selected = X_train[selected_features_sfs]
X_test_selected = X_test[selected_features_sfs]


regressor = LinearRegression()
regressor.fit(X_train_selected, y_train)
# Prediction
y_pred = regressor.predict(X_test_selected)
# mean squared error
mse_sfs = mean_squared_error(y_test, y_pred)

print("MSE:", mse_sfs)

# Compute the mean absolute error
mae_sfs = mean_absolute_error(y_test, y_pred)
print('MAE: {:.2f}'.format(mae_sfs))

# Compute the root mean squared error
rmse_sfs = mean_squared_error(y_test, y_pred, squared=False)
print('RMSE: {:.2f}'.format(rmse_sfs))

r_sq_sfs = regressor.score(X_train_selected, y_train)
print('R^2: {:.2f}'.format(r_sq_sfs))

print("intercept: ", regressor.intercept_)

print("slope: ", regressor.coef_)
print("Variance score: ", regressor.score(X_test_selected, y_test))
```

```python
# RFE
rfe = RFE(estimator=DecisionTreeClassifier(), n_features_to_select=10)
rfe.fit(X_train, y_train)
```

```python
features = X_train.columns.to_list()    
selected_features_rfe = []
for i in range(X.shape[1]):
    if rfe.support_[i]:
        print('Column: %d, %s, Selected %s, Rank: %.3f' % (i, features[i], rfe.support_[i], rfe.ranking_[i]))
        selected_features_rfe.append(features[i])
```

```python
X_train_selected = X_train[selected_features_rfe]
X_test_selected = X_test[selected_features_rfe]


regressor = LinearRegression()
regressor.fit(X_train_selected, y_train)
# Prediction
y_pred = regressor.predict(X_test_selected)
# mean squared error
mse_rfe = mean_squared_error(y_test, y_pred)

print("MSE:", mse_rfe)

# Compute the mean absolute error
mae_rfe = mean_absolute_error(y_test, y_pred)
print('MAE: {:.2f}'.format(mae_rfe))

# Compute the root mean squared error
rmse_rfe = mean_squared_error(y_test, y_pred, squared=False)
print('RMSE: {:.2f}'.format(rmse_rfe))

r_sq_rfe = regressor.score(X_train_selected, y_train)
print('R^2: {:.2f}'.format(r_sq_rfe))

print("intercept: ", regressor.intercept_)

print("slope: ", regressor.coef_)
print("Variance score: ", regressor.score(X_test_selected, y_test))
```

```python
# categories
qc_order = ['Better', 'Average', 'Worse']
cc_order = ['Lower', 'Average', 'Higher']


# Convert to ordered categorical variables
df['Procedure.Heart Attack.Quality'] = pd.Categorical(df['Procedure.Heart Attack.Quality'], categories=qc_order, ordered=True)
df['Procedure.Heart Attack.Value'] = pd.Categorical(df['Procedure.Heart Attack.Value'], categories=cc_order, ordered=True)



# Pivot to create a matrix
matrix = df.pivot_table(index='Procedure.Heart Attack.Quality', columns='Procedure.Heart Attack.Value', values='Procedure.Heart Attack.Cost')



# font scale and color palette for the plot
sns.set(font_scale=1.2)
sns.set_palette("Blues")


#  plot
sns.heatmap(matrix, annot=True, cmap='PuBuGn')
# labels and title
plt.xlabel('comparison to national Cost of care')
plt.ylabel('comparison to national Quality of care')
plt.title("Cost of Heart Attack procedure by Quality and Cost of Care")

plt.show()
```
